# KNN Classification

```python
# Importing the required libraries

import numpy as np
import pandas as pd
import os
```

```python
# For test-train splitting, importing the function from sklearn
from sklearn.model_selection import train_test_split
```

```python
#Reading the file from the specified folder
os.chdir('C:\\Users\\Manoj\\IML')
iris = pd.read_csv('IRIS.csv')
iris
```

```python
#Seperating the target and predictor variables

X = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
#y = iris[['species']]
y = iris.species
```

```python
# Stratify is to divide in homogenous

(X_Train, X_Test, y_Train, y_Test) = train_test_split(X, y, test_size = 0.3, stratify = y, random_state = 100)
```

```python
#Importing the KNN Classifier function from sklearn library
from sklearn.neighbors import KNeighborsClassifier

#Instantiate Learning model (k=3)
knn = KNeighborsClassifier(n_neighbors=3) #3NN

#Fitting the model
knn.fit(X_Train, y_Train)
```

```python
#Predicting the response
pred = knn.predict(X_Test)
```

```python
#Converting it into a dataframe
pd.DataFrame(pred)
```

# KNN Regression

```python
#Reading the file from the specified folder

os.chdir('C:\\Users\\Manoj\\IML')
cars = pd.read_csv('cars.csv')
cars
```

```python
#Seperating the target and predictor variables

X = cars[['Displacement', 'Horsepower', 'Weight', 'Acceleration']]
#y = iris[['species']]
y = cars.MPG
```

```python
# Stratify is to divide in homogenous

(X_Train, X_Test, y_Train, y_Test) = train_test_split(X, y, test_size = 0.3, random_state = 7)
```

```python
#Importing the Kneighbours Regressor function from sklearn package
from sklearn.neighbors import KNeighborsRegressor

#Instantiate Learning model (k=3)
knn = KNeighborsRegressor(n_neighbors=3) #3NN

#Fitting the model
knn.fit(X_Train, y_Train)
```

```python
#Predicting the response
pred = knn.predict(X_Test)
```

```python
#Converting into a dataframe
pd.DataFrame(pred)
```
