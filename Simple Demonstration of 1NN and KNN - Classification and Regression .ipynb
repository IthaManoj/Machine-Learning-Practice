```python
#Importing the required libraries

import os
import numpy as np
import pandas as pd
```

```python
#Defining the euclidean function for calculating euclidean distance

def euclidian(a, b):
    d = np.sqrt(sum((a-b)**2))
    return d
```

## 1NN algorithm for Regression

```python
#Reading the cars.csv file from the folder

data = pd.read_csv('cars.csv')
data
```

```python
#Taking the data to be trained from the iris dataset with the help of iloc

#Predicted Variables
X_Train = data.iloc[:301, 3:6]

#Taking the data to be tested from the iris dataset with the help of iloc

#Predicted Variables
X_Test = data.iloc[301:406, 3:6]
```

```python
#Taking the data to be trained from the iris dataset with the help of iloc

#Target Variable
Y_Train = data.iloc[:301, 1]

#Taking the data to be trained from the iris dataset with the help of iloc

#Target Variable
Y_Test = data.iloc[301:406, 1]
```

```python
X_Train_data_Mean = X_Train.mean()
X_Train_data_Std = X_Train.std()
```

```python
#Standardizing the training data with the help of the formula

# (Data - Mean)/Standard Deviation

X_Train = (X_Train - X_Train_data_Mean)/X_Train_data_Std
X_Train
```

```python
#Standardizing the training data with the help of the formula

# (Data - Mean)/Standard Deviation

X_Test = (X_Test - X_Train_data_Mean)/X_Train_data_Std
X_Test
```

```python
def One_NN_Regression(X_train, x):
    
    #Taking an empty list to store the distances
    dist = []
    
    #Taking an empty list to store the predicted values
    Y_Pred = []

    for i in range(0, len(x)):
        
        for j in range(0, len(X_train)):
            
            #Calling the euclidean distance function to calculate the distances between trained and test data of predicted variables
            d = euclidian(X_train.iloc[j,:], x.iloc[i,:])
            
            #Appending the distances we get in to the dist list
            dist.append(d)
       
        #Calculating the least distance we obtained with the help of indexes
        min_distance = dist.index(min(dist))
        
        #Appending the minimum distances we obtained for individual test data in to the predicted list
        Y_Pred.append(Y_Train.iloc[min_distance,])
        
         #Making the list empty again
        dist  = []
    
    return pd.DataFrame(Y_Pred, Y_Test)       
```

```python
One_NN_Regression(X_Train, X_Test)
```

## KNN Algorithm for Regression 

```python
def KNN_Regression(x_train, x, k):
    
    #Taking an empty list to store the distances
    dist = []
    
    #Taking an empty list to store the predicted values
    Y_Pred = []
    
    for i in range(0, len(x)):
        
        for j in range(0, len(x_train)):
            
            #Calling the euclidean distance function to calculate the distances between trained and test data of predicted variables
            d = euclidian(x_train.iloc[j,:], x.iloc[i,:])
            
            #Appending the distances we get in to the dist list
            dist.append(d)
        
        #Creating a dictionary of distances for every instance of training data to sort the distances in ascending order to get K Nearest neighbours 
        df_dist = {"Index":np.arange(0,len(x_train)), "Distance": dist }
        
        #Converting the dictionary into Data Frame
        df_dist = pd.DataFrame(df_dist)
        
        #Sorting the values by Distance
        df_dist = df_dist.sort_values(by = "Distance")
        
        #Getting the indexes of the K Nearest Neighbours
        indices = df_dist.iloc[0:k,0]
        
        #Finding the predicted value by taking mean of the MPG values
        min_distance = round(Y_Train.iloc[indices, ].mean(), 1)
        
        #Appending the minimum distances we obtained for individual test data in to the predicted list
        Y_Pred.append(min_distance)
        
        #Making the list empty again
        dist = []
    
    #Creating a dictonary so that we can easy access the test data and the predicted data
    final_result = {"Predicted": Y_Pred, "Actual": Y_Test}
    
    #
    final_result = pd.DataFrame(final_result)
    
    return final_result
```

```python
KNN_Regression(X_Train, X_Test, 4)
```
