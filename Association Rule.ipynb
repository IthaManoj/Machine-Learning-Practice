```python
#Importing the required libraries

import os
import pandas as pd
import numpy as np
```

```python
#Importing the apriori algo and association rules functions from the mlxtend library

from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#Importing the preprocessing module for transaction encoder from the mlxtend library

from mlxtend.preprocessing import TransactionEncoder
```

```python
#Reading the file from the specified folder

os.chdir('C:\\Users\\Manoj\\IML')
data = pd.read_csv('lastfm.csv')
data
```

```python
#Making a copy of the data

temp = data.copy()
```

```python
#Grouping the data by user and Converting that to a list

#using the tolist() to convert the artists to a list which is stored in the variable a as 'artist'
txns = [a[1]['artist'].tolist() for a in list(temp.groupby(['user']))]

#Initializing the TransactionEncoder() function
txnEnco = TransactionEncoder()

#Using this object to transform this data of 'artists' to a transaction data of binary values
txns_array = txnEnco.fit(txns).transform(txns)

#converting this array of binary values into a dataframe with column names
txns1 = pd.DataFrame(txns_array, columns=txnEnco.columns_)
```

```python
#to generate frequent itemsets using apriori algorithm
freq_itemsets = apriori(df=txns1,min_support=0.01,use_colnames=True)

#Printing the frequent item sets
freq_itemsets
```

```python
#Applying association rule and imposing min confidence threshold as 60 percentage

rules = association_rules(freq_itemsets,min_threshold=0.6)
```

```python
#Displaying sorted rules in ascending order of lift value

rules.sort_values(by='lift')
```
