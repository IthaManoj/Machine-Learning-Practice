```python
#Importing the functions: 
import numpy as np
import pandas as pd
import os                                    
```

```python
# For test-train splitting, importing the function from sklearn library
from sklearn.model_selection import train_test_split

# For finding mean squared error, importing the function from sklearn library
from sklearn.metrics import mean_squared_error 

# Importing the KNN Regressor function from sklearn library
from sklearn.neighbors import KNeighborsRegressor

from math import sqrt  
```

```python
#Reading the file from the specified folder

os.chdir('C:\\Users\\Manoj\\IML')
cars = pd.read_csv('cars.csv')
cars
```

```python
# Randomizing the data
cars = cars.sample(len(cars))

#K value in K fold
k = 5

# Size for each part
split_size = int(len(cars)/k)
split_size
```

```python
# List to store the data 

cars_list = []

#Splitting of the data into k parts and putting in the above declared list
z=0
for i in range(0,k):
    
    #Splitting the data into subsets
    cars_fold = cars.iloc[z : z + split_size] 
    #Increasing the value of z
    z = z + split_size
    #Appending the above divided data into the list
    cars_list.append(cars_fold)
cars_list
```

```python
# total elements in the list should be equal to k
print(len(cars_list))

# total observations in per element of list should be equal to split_size               
for i in range(5):    
    print(len(cars_list[i]))
```

```python
#Defining the KNN Regression function
def Knn_regression(test,train):
    
    #Seperating the target and predictor variables

    X = cars[['Displacement', 'Horsepower', 'Weight', 'Acceleration']]
    #y = iris[['species']]
    y = cars.MPG
    
    # Stratify is to divide in homogenous
    (X_Train, X_Test, y_Train, y_Test) = train_test_split(X, y, test_size = 0.3, random_state = 7)
    
    #Instantiate Learning model (k=3)
    knn = KNeighborsRegressor(n_neighbors=3) #3NN

    #Fitting the model
    knn.fit(X_Train, y_Train)
    
    #Predicting the response
    pred = knn.predict(X_Test)
    
    #Calculating the root mean square error
    error = sqrt(mean_squared_error(y_Test,pred)) 
    return(error)
```

```python
fold = 0
rmse=[]
trained_data = [] 
for fold in range(k):
    fold_select=[]
    
    # for loop will hold 1 part of test data and other parts of combined train data
    for j in range(k):             
        if j != fold:
            
            #will hold the parts of train data
            fold_select.append(j)            
        else:
            test_data = cars_list[j]
    trained_data = pd.concat([cars_list[fold_select[0]],cars_list[fold_select[1]],cars_list[fold_select[2]],cars_list[fold_select[3]]])
        
    err = Knn_regression(test_data,trained_data)
    rmse.append(err)

print(rmse)

mean_rmse = np.mean(rmse)   
print("Mean RMSE is : ", mean_rmse)
```
