# Hill Descent Algorithm - Simple Demonstration

```python
#Importing the required libraries

import matplotlib.pyplot as plt
import numpy as np
```

```python
#Creating the function

def f(x):
    return(4*x**2-38*x+98)
```

```python
#Derivative of the function

def df(x):
    return(8*x-38)
```

```python
#Setting the initial value of x
x_ini = -10

#Marking the initial value on the plot
p = np.arange(-10,20,0.25)

#plotting the graph
plt.plot(p, f(p), '-')   # '-' is used for defining the shape of the curve. we can use '*', '+', '_' also. By default it is '-'

#for drawing the vertical line on the plot
plt.axvline(x_ini, color = 'r', linewidth = 2)

#for displaying the plot
plt.show()
```

## Hill-Descent Algorithm to find the Minimum

```python
import time
```

```python
def hill_descent(x, alpha, tol):
    #x = -10
    
    #initiating the counter
    count = 0
    
    #alpha = 0.2
    #tol = 1e-10 #Tolerance

    while np.abs(df(x)) > tol:
        
        #incrementing the counter
        count = count + 1
        
        #Learning rate function
        x = x - alpha * df(x)
        #time.sleep(1)
        
        #Setting the axis limits
        p = np.arange(-10,20,0.25)
        
        #for plotting the graph
        plt.plot(p, f(p), '-')
    
        #for getting a vertical line on the graph
        plt.axvline(x, color = 'r', linewidth = 1)

    plt.show()
        
    return(x, count)
    #return(count)  #gives the no. of iterations
```

```python
hill_descent(-10,0.2,1e-10)
```

## Animation of Hill Descent Algorithm

```python
import time
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as plt2
import matplotlib.animation as animation 

x = -10
count = 0
alpha = 0.2
tol = 1e-10 #Tolerance

#For getting an emoty figure
fig = plt.figure()

ax = fig.add_subplot(111)

while np.abs(df(x)) > tol:
    
    #Incrementing the counter
    count = count + 1
    
    #Learning rate function
    x = x - alpha * df(x)
    #time.sleep(1)
    
    #Setting the axis limits
    p = np.arange(-10,20,0.25)
    
    #for plotting the graph
    plt.plot(p, f(p), '-')
    
    #for getting a vertical line on the graph
    plt.axvline(x, color = 'r', linewidth = 1)
    
    #used to show the animation by drawing lines
    fig.canvas.draw()
    
    #Time interval between each step
    time.sleep(0.5)
    fig.canvas.flush_events()
    
    
    
print('Minimum value of f(x): ', x)
print('Number of iteartions reuqired to get the minimum value: ', count)  #gives the no. of iterationss
```
