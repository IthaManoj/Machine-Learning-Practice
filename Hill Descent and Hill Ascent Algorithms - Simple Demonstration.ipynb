# Hill Descent Algorithm - Simple Demonstration

```python
#Importing the required libraries

import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
#Creating the function

def f(x):
    return(4*x**2-38*x+98)
```

```python
#Derivative of the function

def df(x):
    return(8*x-38)
```

```python
def hill_descent(x, alpha, tol):
    #x = -10
    
    #initiating the counter
    count = 0
    
    #alpha = 0.2
    #tol = 1e-10 #Tolerance

    while np.abs(df(x)) > tol:
        
        #incrementing the counter
        count = count + 1
        
        #Learning rate function
        x = x - alpha * df(x)
        #time.sleep(1)
        
        #Setting the axis limits
        p = np.arange(-10,20,0.25)
        
        #for plotting the graph
        plt.plot(p, f(p), '-')
    
        #for getting a vertical line on the graph
        plt.axvline(x, color = 'r', linewidth = 1)

    plt.show()
        
    return(x, count)
    #return(count)  #gives the no. of iterations
```

```python
hill_descent(-10,0.2,1e-10)
```

# Hill Ascent Algorithm - Simple Demonstration

```python
#Creating the function

def f1(x):
    return(-4*x**2+38*x-98)
```

```python
#Derivative of the function

def df1(x):
    return(-8*x+38)
```

```python
def hill_ascent(x, alpha, tol):
    #x = -10
    
    #initiating the counter
    count = 0
    #alpha = 0.2
    #tol = 1e-10 #Tolerance

    while np.abs(df1(x)) > tol:
        
        #incrementing the counter
        count = count + 1
        
        #Learning rate function
        x = x + alpha * df1(x)
        #time.sleep(1)
        
        #Setting the axis limits
        p = np.arange(-10,20,0.25)
        
         #for plotting the graph
        plt.plot(p, f1(p), '-')
        
        #for getting a vertical line on the graph
        plt.axvline(x, color = 'r', linewidth = 1)
    
    #for showing the graph
    plt.show()
        
    return(x, count)
    #return(count)  #gives the no. of iterations
```

```python
hill_ascent(-8,0.2,1e-4)
```

## Hill-Ascent Algorithm to find the Maximum
